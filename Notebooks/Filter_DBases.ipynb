```python
from rdkit import Chem
import os
```

```python
amines=Chem.SmilesWriter('amines.smiles')
isocyanide=Chem.SmilesWriter('isocyanide.smiles')
acids=Chem.SmilesWriter('acids.smiles')
aldehyde=Chem.SmilesWriter('aldehydes.smiles')
for file in os.listdir('../ZINC_fragments/'):
    if '.smi' in file and not '.smiles' in file:
        print (file)
        mols=Chem.SmilesMolSupplier(file)
        for mol in mols:
            if mol.HasSubstructMatch(Chem.MolFromSmarts('[NX3;H2;!$(NC=[!#6]);!$(NC#[!#6])][#6]'))==True:
                amines.write(mol)
            if mol.HasSubstructMatch(Chem.MolFromSmarts('[CX3](=O)[OX2H1]'))==True:
                acids.write(mol)
            if mol.HasSubstructMatch(Chem.MolFromSmarts('[CX3H1](=O)[#6]'))==True:
                aldehyde.write(mol)
            if mol.HasSubstructMatch(Chem.MolFromSmarts('[$([#1,*])][N+]#[C-]'))==True:
                isocyanide.write(mol)
amines.close()
isocyanide.close()
acids.close()
aldehyde.close()
```
